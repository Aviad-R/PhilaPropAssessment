The Philadelphia City Council sets the property tax rate yearly and in 2020 the rate was set at 1.3998%, unchanged from fiscal year 2019 (see the table below for recent historical rates). City Council has enacted numerous policies to abate property values for various economic and political reasons.

The 10-year abatement of Real Estate Taxes, enacted in 1997, exempts from tax the added value from new construction or rehabilitation of both residential and commercial properties, allowing the property owner to pay no tax on improvements for a 10-year period. Other types of property abatements and exemptions exists for non-profits, long time homeowner residents and disabled persons.  The 10-year tax abatement is most significant, impactful and politically consequential abatement due to the current state of the Philadelphia real estate market.

Our goal is to explore, analyze and investigate Philadelphia Property Assessment data that is available via [OpenDataPhilly.org](OpenDataPhilly.org). We are going to attempt to answer questions that are relevant to Philadelphia’s citizens, property owners, elected officials and real estate professionals using methods and techniques that are learned from DSCI521. Analyses will be coded in Python and its associated libraries using Jupyter notebooks.

### INTRODUCTION

**Analysis Goals**

By analyzing the Philadelphia Property Assessment data we are looking to assess the effect(s) of the 10-year tax abatement policy. The City of Philadelphia has previously hired outside firms to assess the economic impact of the abatement, most recently in 2014 and 2018, we look to confirm and build off their conclusions. Additionally, we have posed our own questions of interest below that will guide us further in our analysis of the data

**Interested Parties**

This analysis would be of interest to real estate professionals, residents, property owners and taxpayers in Philadelphia, policy makers, elected officials. The City and School District of Philadelphia have the most significant stake in this analysis as the benefits and disadvantages of the 10-year tax abatement policy go directly towards the funding for the School District.

**Application and Investigation**

This analysis could help homeowners and the City officials with more accurate assessments. Homeowners would be able to use statistical evidence to appeal their property values, potentially saving them money on their property taxes. This could help city officials identify property owners who are non-compliant with the requirement of particular abatements. 

**Data Availability**

Data is available through [OpenDataPhilly.org](https://www.opendataphilly.org/). Data exists for property assessments from 2014-2020 as well as details of each individual property. However, the data is limited in that it is static and stale, whereas, real estate is a dynamic asset and values can change based on micro and macro economic trends. Further, the data will need to be cleaned and any assumptions we take in cleaning the data has the potential to impact the results of the analysis. 

**Continued Analysis**

With more time and resources we could leverage other public and private data sets to make a more robust analysis. Having access to financial information would allow us to more accurately and deeply assess the financial impact of the abatement and whether modifying or cancelling the abatement program would be beneficial to the City and its citizens.

### Questions of Interest

- How many properties are abated? What is the revenue impact to the city and school district?

- What is the distribution across property type (commercial vs. residential) of the abatement in terms of counts and dollar amount.

- Can we map abated properties over time to identify areas of the city that are experiencing development booms (are these areas the hot neighborhoods - Northern Liberties, Kensington, Point Breeze, Brewerytown)

- Can we use cosine similarity to find properties that are out of line with the market (either underassessed or overassessed based on numerous variables - bedrooms, baths, square footage, year built, etc.)

- Can we use historical data to predict areas of the city that we would suspect future development to occur.

- Analyze the age of the residential housing stock (statistics on the age of homes, whether completely new development is accounting for the abatement or renovation to existing homes).

- Are individuals or companies taking advantage of the 10 year abatement. Are there instances where individuals have multiple abatements? Homestead abatements are given only to individuals who live in the home so people with more than one abatement maybe non-compliant.

- Can we use regression modeling to more accurately assess homes based on some or all of the variables that are captured by the city assessor. For example can we take the assessments from year 2014-2019, build a model to predict home values for 2020 and compare against actual results to test the accuracy of our model.

- Are properties that have been recently sold more likely to experience assessment increases/decreases based on the properties sale price.
